```python
import pandas as pd

# Load datasets
train_df = pd.read_csv("Training.csv")
desc_df = pd.read_csv("symptom_Description.csv")
prec_df = pd.read_csv("symptom_precaution.csv")

# Get disease columns
train_disease = set(train_df["prognosis"].str.strip().str.lower())

desc_disease = set(desc_df["Disease"].str.strip().str.lower())

prec_disease = set(prec_df["Disease"].str.strip().str.lower())

# ==========================================
# Compare Diseases
# ==========================================

print("Training Dataset Diseases:", len(train_disease))
print("Description Dataset Diseases:", len(desc_disease))
print("Precaution Dataset Diseases:", len(prec_disease))

# ==========================================
# Check Missing Diseases
# ==========================================

print("\nMissing in Description Dataset:")
print(train_disease - desc_disease)

print("\nMissing in Precaution Dataset:")
print(train_disease - prec_disease)

print("\nExtra Diseases in Description Dataset:")
print(desc_disease - train_disease)

print("\nExtra Diseases in Precaution Dataset:")
print(prec_disease - train_disease)

# ==========================================
# Final Check
# ==========================================

if (
    train_disease == desc_disease ==
    prec_disease
):
    print("\n✅ All datasets contain same disease names")
else:
    print("\n❌ Disease names are NOT same in all datasets")
```
